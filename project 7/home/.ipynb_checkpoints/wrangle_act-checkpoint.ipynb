```python
# importing data from Twitter
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer

consumer_key = 'Hidden'
consumer_secret = 'Hidden'
access_token = 'Hidden'
access_secret = 'Hidden'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
```

```python
tweet_ids = rate_df.tweet_id.value_counts()

len(tweet_ids)
```

```python
# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

#### So letโs begin our Data Tidiness steps 

> 1. In the timestamp column we can notice date and time integrated into 1.
.

### Define



We need to separate the timestamp column into date and time


### Code

```python
df_1_clean_14 = df_1_clean_13.copy()
df_1_clean_14['date'] = [d.date() for d in df_1_clean_14['timestamp']]
df_1_clean_14['time'] = [d.time() for d in df_1_clean_14['timestamp']]


df_1_clean_14.head()
```

```python
# now we drop timestamp column
df_1_clean_14.drop(['timestamp'], axis = 1, inplace = True)
```



### Test

```python
df_1_clean_14.head()
```


> 2. We got four different columns like 'puppo', 'doggo', 'floofer', 'pupper' that increases our data size.


### Define


Creating  new column for the variety of the dogs 

### Code

```python
df_1_clean_15 = df_1_clean_14.copy()
df_1_clean_15[['None', 'doggo_dummy']] = pd.get_dummies(df_1_clean_15['doggo'])
df_1_clean_15.drop(['None'], axis = 1, inplace = True)
df_1_clean_15[['None', 'floofer_dummy']] = pd.get_dummies(df_1_clean_15['floofer'])
df_1_clean_15.drop(['None'], axis = 1, inplace = True)
df_1_clean_15[['None', 'pupper_dummy']] = pd.get_dummies(df_1_clean_15['pupper'])
df_1_clean_15.drop(['None'], axis = 1, inplace = True)
df_1_clean_15[['None', 'puppo_dummy']] = pd.get_dummies(df_1_clean_15['puppo'])
df_1_clean_15.drop(['None'], axis = 1, inplace = True)
df_1_clean_15.head()
```

```python

df_1_clean_15['check'] = df_1_clean_15.doggo_dummy + df_1_clean_15.floofer_dummy + df_1_clean_15.pupper_dummy + df_1_clean_15.puppo_dummy 
df_1_clean_15[df_1_clean_15.check >= 2]
```

Now we can see more than one category of dogs refered in the same post

```python
df_1_clean_16 = df_1_clean_15.copy()
df_1_clean_16['type'] = 'None'
df_1_clean_16.loc[((df_1_clean_16["doggo"] != 'None') & (df_1_clean_16["floofer"] == 'None') & (df_1_clean_16["pupper"] == 'None') & (df_1_clean_16["puppo"] == 'None')),"type"]= 'doggo'
df_1_clean_16.loc[((df_1_clean_16["doggo"] == 'None') & (df_1_clean_16["floofer"] != 'None') & (df_1_clean_16["pupper"] == 'None') & (df_1_clean_16["puppo"] == 'None')),"type"]= 'floofer'
df_1_clean_16.loc[((df_1_clean_16["doggo"] == 'None') & (df_1_clean_16["floofer"] == 'None') & (df_1_clean_16["pupper"] != 'None') & (df_1_clean_16["puppo"] == 'None')),"type"]= 'pupper'
df_1_clean_16.loc[((df_1_clean_16["doggo"] == 'None') & (df_1_clean_16["floofer"] == 'None') & (df_1_clean_16["pupper"] == 'None') & (df_1_clean_16["puppo"] != 'None')),"type"]= 'puppo'
df_1_clean_16.loc[((df_1_clean_16["doggo"] != 'None') & (df_1_clean_16["floofer"] != 'None') & (df_1_clean_16["pupper"] == 'None') & (df_1_clean_16["puppo"] == 'None')),"type"]= 'doggo & floofer'
df_1_clean_16.loc[((df_1_clean_16["doggo"] != 'None') & (df_1_clean_16["floofer"] == 'None') & (df_1_clean_16["pupper"] != 'None') & (df_1_clean_16["puppo"] == 'None')),"type"]= 'doggo & pupper'
df_1_clean_16.loc[((df_1_clean_16["doggo"] != 'None') & (df_1_clean_16["floofer"] == 'None') & (df_1_clean_16["pupper"] == 'None') & (df_1_clean_16["puppo"] != 'None')),"type"]= 'doggo & puppo'
df_1_clean_16.drop(['doggo', 'floofer', 'pupper', 'puppo', 'doggo_dummy', 'floofer_dummy', 'pupper_dummy', 'puppo_dummy', 'check'], axis = 1, inplace = True)
```
